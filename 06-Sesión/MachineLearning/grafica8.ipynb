```python
import matplotlib.pyplot as plt
import pandas as pd

# Ruta del archivo
file_path = "D:/Documents/complementario-3125033/06-Sesin/MachineLearning/machineLearning.xlsx"

# Leer el archivo completo
data = pd.read_excel(file_path, header=None)  # No asumimos encabezados, cargamos todo
print("Primeras filas del DataFrame:\n", data.head())  # Verifica cmo se cargaron los datos

# Localizar los encabezados reales
headers_row = data[data.iloc[:, 0] == "Semana"].index[0]  # Encuentra la fila donde empieza "Semana"
data.columns = data.iloc[headers_row]  # Usa esa fila como nombres de columna
data = data[headers_row + 1:].reset_index(drop=True)  # Filtra los datos por debajo de los encabezados
data.columns.name = None  # Limpia el nombre del ndice de columnas

# Verificar los nombres finales de las columnas
print("Columnas detectadas:", data.columns)
print("Primeras filas procesadas del DataFrame:\n", data.head())

# Extraer columnas relevantes
semanas = pd.to_datetime(data['Semana'], errors='coerce')  # Convertir la columna 'Semana' a datetime
machineLearning = pd.to_numeric(data['Aprendizaje automatico'], errors='coerce')  # Aseguramos conversin numrica
python = pd.to_numeric(data['Python'], errors='coerce')

# Graficar los datos
plt.figure(figsize=(10, 6))
plt.plot(semanas, machineLearning, label='Aprendizaje automatico', color='red', marker='o')
plt.plot(semanas, python, label='Python', color='blue', marker='o')

# Personalizacin del grfico
plt.ylim(0, 100)
plt.xticks(semanas[::len(semanas)//10], [str(year) for year in semanas.dt.year[::len(semanas)//10]])  # Ajustar las etiquetas de aos
plt.xlabel('Ao')
plt.ylabel('ndice de Popularidad')
plt.title('Comparativa de Popularidad: Aprendizaje Automatico vs Python')
plt.legend()
plt.grid()

# Mover la leyenda al lado izquierdo
plt.legend(loc='upper left')  # Aqu se especifica la ubicacin de la leyenda

# Mostrar la grfica
plt.tight_layout()
plt.show()
```
